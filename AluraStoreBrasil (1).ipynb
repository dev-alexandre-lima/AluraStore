```python
# üì¶ Importa√ß√£o das bibliotecas necess√°rias
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# üì• Importa√ß√£o dos dados das quatro lojas
url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

# üîó Leitura dos arquivos CSV e armazenamento em uma lista
lojas = [pd.read_csv(url) for url in [url1, url2, url3, url4]]

# üé® Configura√ß√£o do estilo dos gr√°ficos
sns.set(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)

# üìä Lista para armazenar os resultados de cada loja
relatorio = []

# üîç Loop para analisar cada loja individualmente
for i, loja in enumerate(lojas, start=1):

    # 1Ô∏è‚É£ An√°lise do Faturamento: calcula o faturamento total da loja
    loja['faturamento'] = loja['preco'] * loja['quantidade']
    faturamento_total = loja['faturamento'].sum()

    # 2Ô∏è‚É£ Vendas por Categoria: identifica a categoria mais vendida
    categoria_popular = loja['categoria'].value_counts().idxmax()

    # 3Ô∏è‚É£ M√©dia de Avalia√ß√£o das Lojas: calcula a m√©dia das avalia√ß√µes dos clientes
    avaliacao_media = loja['avaliacao'].mean()

    # 4Ô∏è‚É£ Produtos Mais e Menos Vendidos: identifica os produtos com maior e menor quantidade vendida
    produto_mais_vendido = loja.groupby('produto')['quantidade'].sum().idxmax()
    produto_menos_vendido = loja.groupby('produto')['quantidade'].sum().idxmin()

    # 5Ô∏è‚É£ Frete M√©dio por Loja: calcula o custo m√©dio de frete
    frete_medio = loja['frete'].mean()

    # üìã Armazena os resultados em um dicion√°rio
    relatorio.append({
        'Loja': f'Loja {i}',
        'Faturamento Total': round(faturamento_total, 2),
        'Categoria Mais Popular': categoria_popular,
        'M√©dia de Avalia√ß√£o': round(avaliacao_media, 2),
        'Produto Mais Vendido': produto_mais_vendido,
        'Produto Menos Vendido': produto_menos_vendido,
        'Frete M√©dio': round(frete_medio, 2)
    })

# üìë Cria√ß√£o do DataFrame com o relat√≥rio consolidado
df_relatorio = pd.DataFrame(relatorio)

# üìà Gr√°fico 1: Faturamento Total por Loja
plt.figure()
sns.barplot(x='Loja', y='Faturamento Total', data=df_relatorio)
plt.title('Faturamento Total por Loja')
plt.show()

# üìà Gr√°fico 2: M√©dia de Avalia√ß√£o por Loja
plt.figure()
sns.barplot(x='Loja', y='M√©dia de Avalia√ß√£o', data=df_relatorio)
plt.title('M√©dia de Avalia√ß√£o por Loja')
plt.show()

# üìà Gr√°fico 3: Frete M√©dio por Loja
plt.figure()
sns.barplot(x='Loja', y='Frete M√©dio', data=df_relatorio)
plt.title('Frete M√©dio por Loja')
plt.show()

# üìä Exibi√ß√£o do relat√≥rio final com todas as m√©tricas
df_relatorio

```
